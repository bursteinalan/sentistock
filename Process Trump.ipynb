```python
from pathlib import Path
import pandas as pd
```

```python
file = Path('data/tweets/trumpdb.csv')
df = pd.read_csv(file)
df = df.rename(columns={'created_at':'Date', 'retweet_count': 'number_of_shares', 'favorite_count': 'number_of_likes', 'text':'content'})
df['author'] = ['Trump'] * df.shape[0]
df = df[['Date', 'author', 'content', 'number_of_likes', 'number_of_shares']]
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%m-%d-%Y %H:%M:%S')
df = df.sort_values(by='Date')
df = df.reset_index(drop=True)
df = df.fillna(0).astype({'number_of_shares': 'int64'})
df.to_csv(Path('data/in/twitter/Trump.csv'), index=False)
df
```

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("all-the-news.db")
```

```python
cur = conn.cursor()
```

```python
res = cur.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;").fetchall()
```

```python
res
```

```python
table = pd.read_sql_query("SELECT * from longform", conn)
```

```python
table
```

```python
table.groupby('publication').count()
```

```python
interesting_pubs = ['CNN', 'New York Times', 'Washington Post']
```

```python
for pub in interesting_pubs:
    df = table[table['publication']==interesting_pubs]
    df = df[['title','author','date']]
    df.to_csv(pub+'.csv', index=False)
```
